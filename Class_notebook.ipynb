```python
#import libraries
import pandas as pd
import numpy as np
```

```python
data_frame_liquors = pd.read_csv('database_liquors.csv')
```

```python
data_frame_liquors.info()
```

```python
data_frame_liquors.head()
```

```python
len(data_frame_liquors)
```

```python
#renombrando una columna para usar id
data_frame_liquors = data_frame_liquors.rename(columns = {'invoice':'invoice_id'})
```

```python
data_frame_liquors.head()
```

```python
#cambiando columnas a mayusculas
data_frame_liquors.columns = [x.upper() for x in data_frame_liquors.columns]
#hay veces en las que las columnas pueden venir en mayusculas (eso no nos conviene)
```

```python
data_frame_liquors
```

```python
#cambiando columnas a minusculas
data_frame_liquors.columns = [x.lower() for x in data_frame_liquors.columns]
```

```python
data_frame_liquors
```

```python
#convirtiendo la columna date de type:obj a type:pd.date_time
data_frame_liquors['date'].dtypes
```

```python
data_frame_liquors['date'] = pd.to_datetime(data_frame_liquors['date'])
```

```python
data_frame_liquors['date'].dtypes #<M8[ns] = datetime64[ns]
```

```python
np.dtype('datetime64[ns]') == np.dtype('<M8[ns]')
```

```python
data_frame_liquors['date_mes'] = data_frame_liquors['date'].dt.month
```

```python
data_frame_liquors.head()
```

```python
#truncar valores
data_frame_liquors['date_mes_truncate'] = data_frame_liquors['date'].values.astype('datetime64[M]')
```

```python
data_frame_liquors['date_mes_truncate']
```

```python
df_nulo = data_frame_liquors.isnull().sum()
```

```python
#creando dataframe para todos los datos donde el item_description sea Captain Morgan
data_captain_morgan = data_frame_liquors.query('item_description == "Captain Morgan Original Spiced"')
```

```python
#creando dataframe para describir las ventas de cada botella por ciudad
df_summarize = data_frame_liquors.groupby(['item_description','category_name',
                                           'county','date_mes_truncate']).agg({'bottles_sold':'sum', 
                                                                                'sale_usd':'sum',
                                                                              'state_bottle_cost':'mean',
                                                                              'state_bottle_retail':'mean'}).reset_index()
```

```python
df_summarize = df_summarize.set_index('item_description')
```

```python
#Filtrando por category_name donde 'Rum' este dentro de ese registro (Como si fuera findall de regex)
df_rums = data_frame_liquors[data_frame_liquors['category_name'].str.contains('Rum')]
df_rums
```

```python
#Agrupando por cada botella y viendo cuanto se vendio y cuantas se vendieron
df_rums.groupby('item_description')[['sale_usd','bottles_sold']].sum()
```

```python
#Agrupando por botella y viendo cuanto se vendio y cuanto cuesta la botella en promedio
data_frame_liquors[data_frame_liquors['category_name'].str.contains('Rum')].groupby('item_description').agg({'sale_usd':'sum','bottles_sold':'mean'}).sort_values('sale_usd',ascending=False)
```
